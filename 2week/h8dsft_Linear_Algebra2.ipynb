2. Lakukan perkalian terhadap matrix berikut: 

        Jawab:

```python
matrix_A=np.array([[23,50,19],
                  [7,12,109],
                  [57,67,98]])
matrix_B=np.array([[17,
                   22,
                   19]])

print("Membuat Matrix A")
print(matrix_A)
print("")
print("Membuat Matrix B")
print(matrix_B)
print("")
print("Perkalian Matrix A dan B")
print(matrix_B*matrix_A)
```

3. Lakukan Transpose, hitung determinan, dan inverse 
    terhadap matrix dibawah ini:

        Jawab:

```python
matrix_A=np.array([[23,50,19],
                  [7,12,109],
                  [57,67,98]])

print('Matrix A :')
print(matrix_A)

print('Transpose:')
print(matrix_A.T)
print('')

print('Inverse :')#Inverse
np.linalg.inv(matrix_A.T)
```

```python
print('Determinan :')#determinan
np.linalg.det(matrix_A.T)
```
